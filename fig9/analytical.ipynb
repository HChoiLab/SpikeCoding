```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.patches as mpatches
from matplotlib import font_manager
```

```python
font_dirs = ['/storage/home/hcoda1/8/zmobille3/scratch/arial']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
for font_file in font_files:
    font_manager.fontManager.addfont(font_file)
# font_manager.get_font_names()

plt.rcParams['font.family']='Arial'
plt.rcParams['font.size']=18
plt.rcParams['axes.titlesize']=18
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
```

```python
nbinvec=np.arange(1,21,1)
count_slopes=[np.log2(nbins+1) for nbins in nbinvec]
time_slopes=[nbins for nbins in nbinvec]
plt.plot(nbinvec,count_slopes,label='spike count',color='lightseagreen',marker='o')
plt.plot(nbinvec,time_slopes,label='spike timing',color='purple',marker='o')
plt.xlim(0.5,20.5)
plt.ylim(0,21)
plt.xticks([1,5,10,15,20])
# plt.legend()
plt.xlabel(r'no. time bins  $n_{bins}=T/\tau_{ref}$')
plt.ylabel(r'slope of $\widetilde{H}$ v.s. $N_{nrn}$ (bits/nrn)')
sns.despine()
plt.savefig('slope_v_nbins.png',bbox_inches='tight',dpi=200)
plt.savefig('slope_v_nbins.svg',bbox_inches='tight',dpi=200)
plt.show()
```

```python
T=15 # response window
tauref=5 # refractory period
nbins=int(T/tauref) # number of time bins

Nnrnvec=np.arange(1,21,1)

count_entropy_bound=np.array([Nnrn*np.log2(nbins+1) for Nnrn in Nnrnvec])
time_entropy_bound=np.array([Nnrn*nbins for Nnrn in Nnrnvec])

plt.plot(Nnrnvec,count_entropy_bound,color='lightseagreen',marker='o')
plt.plot(Nnrnvec,time_entropy_bound,color='purple',marker='o')
plt.plot(Nnrnvec,30*np.ones(len(Nnrnvec)),linestyle='--',color='gray')
# print(count_entropy_bound)
# print(time_entropy_bound)
count_idx = np.argwhere(count_entropy_bound==30.)[0]
count_Nnrn = Nnrnvec[count_idx]
time_idx = np.argwhere(time_entropy_bound==30)[0]
time_Nnrn = Nnrnvec[time_idx]
plt.xlim(0.5,20.5)
plt.ylim(0,63)
plt.xticks([1,5,10,15,20])
plt.vlines(count_Nnrn,0,30,linestyle='--',color='lightseagreen')
plt.vlines(time_Nnrn,0,30,linestyle='--',color='purple')
plt.ylabel(r'entropy upper bound $\widetilde{H}$ (bits)')
plt.xlabel('no. neurons $N_{nrn}$')
sns.despine()
plt.savefig('ent_v_Nnrn.png',bbox_inches='tight',dpi=200)
plt.savefig('ent_v_Nnrn.svg',bbox_inches='tight',dpi=200)
plt.show()
```

```python
T=15 # response window
tauref=5 # refractory period
nbins=int(T/tauref) # number of time bins

Nnrnvec=np.arange(1,21,1)

count_entropy_bound=np.array([Nnrn*np.log2(nbins+1) for Nnrn in Nnrnvec])
time_entropy_bound=np.array([Nnrn*nbins for Nnrn in Nnrnvec])

count_entropy_rate = np.array([x/T for x in count_entropy_bound])
time_entropy_rate = np.array([x/T for x in time_entropy_bound])

plt.plot(Nnrnvec,count_entropy_rate,color='lightseagreen',marker='o')
plt.plot(Nnrnvec,time_entropy_rate,color='purple',marker='o')
plt.vlines(10,0,60/T,linestyle='--',color='gray')

count_idx = np.argwhere(count_entropy_bound==20.)[0]
count_Nnrn = Nnrnvec[count_idx]
time_idx = np.argwhere(time_entropy_bound==30.)[0]
time_Nnrn = Nnrnvec[time_idx]
plt.xlim(0.5,20.5)
plt.ylim(0,63/T)
plt.xticks([1,5,10,15,20])

countRateToPrint = count_entropy_rate[count_idx]
timeRateToPrint = time_entropy_rate[time_idx]
print(timeRateToPrint[0])

timeString = f'{timeRateToPrint[0]} bits/ms'
plt.text(4.5,32/T,timeString,color='purple')

countString = f'{countRateToPrint[0]:.2} bits/ms'
plt.text(10.5,15.5/T,countString,color='lightseagreen')
# plt.hlines(20,0,10,linestyle='--',color='lightseagreen')
# plt.hlines(30,0,10,linestyle='--',color='purple')
plt.ylabel(r'entropy rate $\frac{\widetilde{H}^{pop}}{T}$ (bits/ms)')
plt.xlabel('no. neurons $N_{nrn}$')
sns.despine()
plt.savefig('entRate_v_Nnrn.png',bbox_inches='tight',dpi=200)
plt.savefig('entRate_v_Nnrn.svg',bbox_inches='tight',dpi=200)
plt.show()
```
